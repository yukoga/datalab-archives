### Model context defined for the notebook
If the notebook deals only one model (maybe multiple versions), you can use %%ml model to define a default model. In other cells, you don't need to specify the model again and it will pick the default model.

```python
%%ml model --name census
```

### Check details of a model

```python
%%ml describe --cloud
```

### Check details of a model version

```python
%%ml describe -v v1 --cloud
```

### Compare multiple versions

```python
%%ml analyze -v v1 v2 --cloud
```
